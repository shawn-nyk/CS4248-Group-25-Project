The core dataset contains 50,000 reviews split evenly into 25k train
and 25k test sets. The overall distribution of labels is balanced (25k
pos and 25k neg).

```python
def files_to_list(path):
    corpus = []
    for filename in os.listdir(path):
        with open(os.path.join(path, filename), 'r') as f: 
            for line in f:
                curr = line.split()
                for w in curr:
                    if w not in stop and 'br' not in w and '\'' not in w:
                        corpus.append(w)
            #    break
            #break
        #break
    return corpus
```

```python
pos_corpus = files_to_list(train_pos_path)
neg_corpus = files_to_list(train_neg_path)
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
stopwords.add('movie')
stopwords.add('film')
stopwords.add('one')
def show_wordcloud(data):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=100,
        max_font_size=100,
        scale=3,
        random_state=1)
   
    wordcloud=wordcloud.generate(" ".join(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')

    plt.imshow(wordcloud)
    plt.show()

```

```python
show_wordcloud(pos_corpus)
```

```python
show_wordcloud(neg_corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

```python
top_n_bigrams=get_top_ngram(pos_corpus,2)[:10]
x,y=map(list,zip(*top_n_bigrams)) 
sns.barplot(x=y,y=x)
```

```python
top_n_bigrams=get_top_ngram(neg_corpus,2)[:10]
x,y=map(list,zip(*top_n_bigrams)) 
sns.barplot(x=y,y=x)
```
